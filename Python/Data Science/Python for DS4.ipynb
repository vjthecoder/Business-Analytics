```python
plt.plot(x,y, label='Sample Sin(x) plot')
plt.title('My Fist plot')
plt.xlabel('Input')
plt.ylabel('Ouptput')
plt.grid(True)
plt.figtext(0.995, 0.01, 'Footnote', ha='right', va='bottom')
plt.legend(loc='best', framealpha=0.5, prop={'size':'small'})
plt.tight_layout(pad=1)
plt.savefig('filename.png')
```

```python
# Plotting in Object Oriented way
x = np.linspace(0,20,1000)
y = np.sin(x)

fig = plt.figure(figsize=(8,4))

ax = fig.add_subplot(1,1,1)
ax.plot(x, y, 'b-', linewidth=3, label='Sample label')
ax.set_ylabel('y axis label', fontsize=14)
ax.set_xlabel('x axis lable', fontsize=14)
ax.legend(loc='best')
ax.grid(True)

fig.suptitle('Sample Plot Title')
fig.tight_layout(pad=1)
fig.savefig('filename2.png', dpi=125)
```
